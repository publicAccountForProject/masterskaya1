```python
#Выведем гистограмму и распределение для каждого из значений относящихся к числовым значеним
#Иначе построим график для категориальных значений

def display_distribution_numeric(data):
    #1 строка, 2 колонки, размер графика
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))

    # Гистограмма слева, на выход получает df, строит график в левой колонке, добавляет линию
    sns.histplot(data, ax=axs[0], kde=True)
    axs[0].set_xlabel('Значение')
    axs[0].set_ylabel('Частота')
    axs[0].set_title('Распределение')

    # Boxplot, на вход получает df, строит график в правой колонке
    sns.boxplot(data, ax=axs[1])
    axs[1].set_xlabel('Значение')
    axs[1].set_title('Boxplot с выбросами')
    #для отступов и построение графика
    plt.tight_layout()
    plt.show()

def display_distribution_categorical(data):
    #построим график для категориальных значений
    fig, axs = plt.subplots(1, 2, figsize=(8, 6))
 
    sns.countplot(data, ax=axs[0])
    axs[0].set_xlabel('Значение')
    axs[0].set_ylabel('Количество')
    axs[0].set_title('Подсчет значений')

    # Boxplot для категориальных данных не строим
    axs[1].axis('off')  # удалим пустой subplot

    plt.tight_layout()
    plt.show()
                
def display_column_distributions(df):
    for column in df.columns:
        
        if column == 'vin' or column == 'seller'  or  column == 'saledate':
            print(f"Колонка с {column} не информативна для вывода. Пропускаем")
            continue
            
        elif df[column].dtype in ['int64', 'float64']:
            if df[column].isnull().any():
                print(f"Колонка {column} содержит NaN значения. Пропускаем")
            else:
                display_distribution_numeric(df[column])
        elif df[column].dtype == 'object':
            if df[column].isnull().any():
                print(f"Колонка {column} содержит NaN значения. Пропускаем")
            else:
                display_distribution_categorical(sorted(df[column]))
        print("Колонка: ", column)
        print("--------------------------------------------------")
```

- Исправление пропусков в make

Заметим, что почти для 2-х процентов одновременно отсутвуют значения в полях make и model, а без них невозможно определить производителя и модель авто для пргноза. Но и удалять строки нельзя, т.к. в kaggle скорее всего будут данные с пропусками. Поэтому используя vininfo по vin заполним make, добавив производителя

```python
# Функция для получения производителя по VIN
def get_manufacturer(vin):
    try:
        vin_obj = Vin(vin)
        return vin_obj.manufacturer
    except:
        return None

# Заполняем пропущенные значения производителя
def insert_manufacturer(df):    
    for index, row in df.iterrows():
        if pd.isnull(row['make']):
            vin = row['vin']
            manufacturer = get_manufacturer(vin)
            df.at[index, 'make'] = manufacturer
```

```python
def cnt_for_rare(df):
    # Подсчет количества значений в каждой группе для категориальных столбцов
    group_counts = df.select_dtypes(include=['object']).apply(lambda x: x.value_counts())

    # Получение списка категориальных столбцов
    columns = [col for col in df.select_dtypes(include=['object']).columns if col not in ('saledate', 'make', 'vin')]
    # Замена значений менее 20 на "rare" для столбцов
    for column in columns:
        rare_values = group_counts[column][group_counts[column] < 20].index
        df[column] = train[column].apply(lambda x: 'rare' if x in rare_values else x)
```

```python
#находим самый частый вариант коробки передач для каждой марки и модели. Им и заполняем пропуски.
def fill_missing_transmission(df):
    # Группировка и заполнение пропущенных значений в зависимости от наличия model
    if 'model' in df.columns:
        df['transmission'] = df.groupby(['make', 'model'])['transmission'].transform(lambda x: x.fillna(stats.mode(x).mode[0]))
    else:
        df['transmission'] = df.groupby(['make'])['transmission'].transform(lambda x: x.fillna(stats.mode(x).mode[0]))

# Пример использования функции
fill_missing_transmission(train)
fill_missing_transmission(test)
```

Вместо пропущенных значений укажем "Unknown" для категориальных столбцов и медиану для числовых столбцов

```python
def fill_missing_values(df):
    # Замена пропущенных значений на медиану для числовых столбцов
    numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
    for column in numeric_columns:
        median_value = df[column].median()
        df[column].fillna(median_value, inplace=True)
    
    # Добавление новой категории "Unknown" для категориальных столбцов
    categorical_columns = df.select_dtypes(include=['object']).columns
    for column in categorical_columns:
        df[column].fillna('Unknown', inplace=True)
```

Как зависит средняя цена от:
- состояния 
- года

```python
def average_price_by(variable = 'condition'):
    # Группировка по состоянию и вычисление средней цены и доверительного интервала
    average_price_by_condition = train.groupby(variable)['sellingprice'].mean()
    confidence_interval = train.groupby(variable)['sellingprice'].agg(lambda x: x.sem() * 1.96)  # Для 95% доверительного интервала

    # Построение графика
    plt.figure(figsize=(8, 6))
    sns.lineplot(x=average_price_by_condition.index, y=average_price_by_condition.values, marker='o')
    plt.fill_between(average_price_by_condition.index, 
                 average_price_by_condition - confidence_interval, 
                 average_price_by_condition + confidence_interval, 
                 color='blue', alpha=0.2)
    plt.title(f'Изменение средней цены в зависимости от {variable}')
    plt.xlabel(variable)
    plt.ylabel('Средняя цена')
    plt.xticks(rotation=0)
    plt.show()
```

```python
#Для прохождения проверки от Яндекс
from sklearn.metrics import make_scorer
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from catboost import CatBoostRegressor
import numpy as np


# функция MAPE 
def custom_mape(y_true, y_pred):
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    return mape

# make_scorer MAPE #https://scikit-learn.org/stable/modules/generated/sklearn.metrics.make_scorer.html
mape_scorer = make_scorer(custom_mape, greater_is_better=False)

# Диапазон гиперпараметров
estimators_range = list(range(1, 10))
max_depth_range = list(range(1, 6)) 


# pipelines для каждой модели отдельно. Делаю так, чтобы видеть лучшие значения Mape для каждой модели. 
#Мне казалось, что тут лучше всего отработает catboost, а оказалось, что GradientBoostingRegressor
rf_pipeline = Pipeline([
    ('random_forest', RandomForestRegressor(n_jobs=-1))
])

catboost_pipeline = Pipeline([
    ('catboost', CatBoostRegressor(random_state=12345, silent=True, n_estimators=100))
])


gradient_boost_pipeline = Pipeline([
    ('gradient_boost', GradientBoostingRegressor(random_state=12345, learning_rate=0.01))
])

#параметры для каждой из модели

param_grid_rf = {
    'random_forest__n_estimators': estimators_range,
    'random_forest__max_depth': max_depth_range
}

param_grid_catboost = {
    'catboost__n_estimators': estimators_range,
    'catboost__max_depth': max_depth_range
}

param_grid_gradient_boost = {
    'gradient_boost__n_estimators': estimators_range,
    'gradient_boost__max_depth': max_depth_range
}

# кросс-валидация GridSearchCV для подбора параметров

grid_search_rf = GridSearchCV(rf_pipeline, param_grid_rf, cv=5, n_jobs=-1, scoring=mape_scorer)
grid_search_catboost = GridSearchCV(catboost_pipeline, param_grid_catboost, cv=5, n_jobs=-1, scoring=mape_scorer)

grid_search_gradient_boost = GridSearchCV(gradient_boost_pipeline, param_grid_gradient_boost, cv=5, n_jobs=-1, scoring=mape_scorer)

# Обучение моделей на тренировочныъ дынных

grid_search_rf.fit(X_train, y_train)
grid_search_catboost.fit(X_train, y_train)
grid_search_gradient_boost.fit(X_train, y_train)

# Лучшая модель и ее параметры

best_rf_model = grid_search_rf.best_estimator_
best_catboost_model = grid_search_catboost.best_estimator_
best_gradient_boost_model = grid_search_gradient_boost.best_estimator_

# Используя лучшие параметры, делаем предсказание на валидационной выборке

rf_predictions = best_rf_model.predict(X_val)
catboost_predictions = best_catboost_model.predict(X_val)
gradient_boost_predictions = best_gradient_boost_model.predict(X_val)
```

```python
# Вычислите MAPE для каждой модели
rf_mape = custom_mape(y_val, rf_predictions)
catboost_mape = custom_mape(y_val, catboost_predictions)

gradient_boost_mape =  custom_mape(y_val, gradient_boost_predictions)

print("MAPE для RandomForestRegressor:", rf_mape)
print("MAPE для CatBoostRegressor:", catboost_mape)
print("MAPE для GradientBoostingRegressor:", gradient_boost_mape)

```
